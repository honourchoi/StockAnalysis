```python
import pandas as pd
import pandas_datareader.data as web
import datetime
import matplotlib.pyplot as plt
import numpy as np
from pandas.plotting import scatter_matrix
from mplfinance.original_flavor import candlestick_ohlc
%matplotlib inline
```

```python
start = datetime.datetime(2020,4,18)
end = datetime.datetime(2021,4,18)
```

```python
google = web.DataReader('GOOGL', 'yahoo', start, end)
```

```python
starbucks = web.DataReader('SBUX', 'yahoo', start, end)
```

```python
tesla = web.DataReader('TSLA', 'yahoo', start, end)
```

```python
ford = web.DataReader('FORD', 'yahoo', start, end)
```

```python
google.head()
```

```python
starbucks.head()
```

```python
google['Open'].plot(label = 'GOOGL Open Price', figsize=(15, 7))
google['Close'].plot(label = 'GOOGL Close Price')
google['High'].plot(label = 'GOOGL High Price')
google['Low'].plot(label = 'GOOGL Low Price')
plt.legend()
plt.title('Google Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
starbucks['Open'].plot(label = 'SBUX Open Price', figsize=(15, 7))
starbucks['Close'].plot(label = 'SBUX Close Price')
starbucks['High'].plot(label = 'SBUX High Price')
starbucks['Low'].plot(label = 'SBUX Low Price')
plt.legend()
plt.title('Starbucks Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
tesla['Open'].plot(label = 'TSLA Open Price', figsize=(15, 7))
tesla['Close'].plot(label = 'TSLA Close Price')
tesla['High'].plot(label = 'TSLA High Price')
tesla['Low'].plot(label = 'TSLA Low Price')
plt.legend()
plt.title('Tesla Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
ford['Open'].plot(label = 'FORD Open Price', figsize=(15, 7))
ford['Close'].plot(label = 'FORD Close Price')
ford['High'].plot(label = 'FORD High Price')
ford['Low'].plot(label = 'FORD Low Price')
plt.legend()
plt.title('Ford Stock Prices')
plt.ylabel('Stock Price')
plt.show()
```

```python
google['Total Traded'] = google['Open'] * google['Volume']
starbucks['Total Traded'] = starbucks['Open'] * starbucks['Volume']
ford['Total Traded'] = ford['Open'] * ford['Volume']
```

```python
google['Total Traded'].plot(figsize=(15, 7))
```
